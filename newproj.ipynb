```python
from sklearn.preprocessing  import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
label_encoder=LabelEncoder()
ds['process_path']=label_encoder.fit_transform(ds['process_path'].astype(str))
ds['sensor_criticality']=label_encoder.fit_transform(ds['sensor_criticality'].astype(str))
ds['alert_type']=label_encoder.fit_transform(ds['alert_type'].astype(str))
ds['@timestamp']=label_encoder.fit_transform(ds['@timestamp'].astype(str))
ds['watchlist_id']=label_encoder.fit_transform(ds['watchlist_id'].astype(str))
ds['tags']=label_encoder.fit_transform(ds['tags'].astype(str))
ds['report_score']=label_encoder.fit_transform(ds['report_score'].astype(str))
ds['alert_severity']=label_encoder.fit_transform(ds['alert_severity'].astype(str))
```
